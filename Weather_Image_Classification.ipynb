```
# import general necessary packages
import pandas as pd
import numpy as np
import os, cv2 

# import the necessary packages -- cnn
from keras import layers
from keras import models
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from keras.preprocessing.image import ImageDataGenerator
from keras.layers.core import Dropout
from keras.layers.core import Dense
from keras.layers import Flatten
from keras.layers import Input
from tensorflow.keras.optimizers import Adam, RMSprop
from keras.callbacks import EarlyStopping
from tensorflow.keras import optimizers

import random as rn
```

## **Model1: CNN Baseline Model**

For a baseline model, we will use the general structure: the convnet will be a stack of alternated Conv2D (with relu activation) and MaxPooling2D layers. Since we’re attacking a multiple-classification problem, we’ll end the network with multiple units (a Dense layer of size 4) and a softmax activation.

```
# Instantiating a small convnet for the weather classification

model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu',
                        input_shape=(150, 150, 3))) # this needs to match with our data prep - upcoming!
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(4, activation='softmax'))
```

Let’s look at how the dimensions of the feature maps change with every successive layer:

```
model.summary() #Viewing model summary
```

For the compilation step, we’ll go with the RMSprop optimizer. Because you ended the network with multiple units, you’ll use categorical_crossentropy as the loss. 

```
# Configuring the model for training
model.compile(loss='categorical_crossentropy',
              optimizer='Adam',
              metrics=['acc'])
```

We will do data preprocessing at this point becasue data should be formatted into appropriately preprocessed floating-point tensors before being fed into the network. We will use ImageDataGenerator to autumatically turn our image files on Google Drive into batches of preprocessed tensors.  

```
# Using ImageDataGenerator to read images from directories
# All images will be rescaled by 1./255
train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        # This is the target directory
        train_dir,
        # All images will be resized to 150x150
        target_size=(150, 150),
        batch_size=20,
        # Since we use categorical_crossentropy loss, we need categorical labels
        class_mode='categorical')

validation_generator = test_datagen.flow_from_directory(
        validation_dir,
        target_size=(150, 150),
        batch_size=20,
        class_mode='categorical')
```

Let’s look at the output of one of these generators:

```
for data_batch, labels_batch in train_generator:
    print('data batch shape:', data_batch.shape)
    print('labels batch shape:', labels_batch.shape)
    break # the generator yields these batches indefinitely if there is no break
```

It yields batches of 150 × 150 RGB images (shape (20, 150, 150, 3)) and categorical labels (shape (20, 4)). There are 20 samples in each batch (the batch size). 

Let’s fit the model to the data using the generator. In order to fit the model, we will use model.fit which expects as its arguments Python generators of train and validation sets that will yield batches of inputs and targets indefinitely. Because the data is being generated endlessly, we should set steps_per_epoch and validation_step to let Keras model know how many samples to draw from the generator before declaring an epoch over.

```
# Fitting the model using a batch generator
import time
start = time.time()

tf.random.set_seed(1)

# early stopping callback
es = EarlyStopping(monitor='val_loss',
                   mode='min',
                   patience=10,
                   restore_best_weights = True)

history = model.fit(
      train_generator,
      steps_per_epoch=55, # ideally, len(trainX) // BS == 1129/20  == 55
      callbacks=[es],
      epochs=100,
      validation_data=validation_generator,
      validation_steps=19)  # ideally, len(valX) // BS == 384/20 == 19 
                            # Total number of steps (batches of samples) to validate before stopping.

end = time.time()
print(end - start)
```

```
# plot the results of the model
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()

plt.figure()

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()
```

We can see that the model is overfitting. Our training accuracy increases linearly over time, until it reaches nearly 100%, while our 
validation accuracy stalls at 70-72%. Our validation loss reaches its minimum after only eight epochs then increases, while the training loss 
keeps decreasing linearly until it reaches nearly 0.

This is because we only have around 1100 training samples which are relatively small . We are now going to use data augmentation to mitigate overfitting.

## **Model2: CNN with Data Augmentation**

We will use data augmentation to generate more training data from existing training samples, by augmenting the samples through a number of random transformations that create a little different images in order that our model would never see the exact same picture twice, thereby getting exposed to more aspects of the data and generalizing better.

```
# Defining a new convnet that includes dropout
model = models.Sequential()
model.add(layers.Conv2D(32, (3, 3), activation='relu',
                        input_shape=(150, 150, 3)))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(64, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Conv2D(128, (3, 3), activation='relu'))
model.add(layers.MaxPooling2D((2, 2)))
model.add(layers.Flatten())
model.add(layers.Dropout(0.5))
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(4, activation='softmax'))

model.compile(loss='categorical_crossentropy',
              optimizer='Adam',
              metrics=['acc'])

# early stopping callback
from keras.callbacks import EarlyStopping
es = EarlyStopping(monitor='val_loss',
                   mode='min',
                   patience=10,
                   restore_best_weights = True)
```

Let's train our network using data augmentation and dropout.

We don't augment the validation data because it is used to evaluate the model, which means it shouldn't be generalized. 

```
# Training the convnet using data-augmentation generators
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True)

# the validation data should not be augmented!
# we simply scale it by the max value (255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        # This is the target directory
        train_dir,
        # All images will be resized to 150x150
        target_size=(150, 150),
        batch_size=20,
        # Since we use categorical_crossentropy loss, we need categorical labels
        class_mode='categorical')

validation_generator = test_datagen.flow_from_directory(
        validation_dir,
        target_size=(150, 150),
        batch_size=20,
        class_mode='categorical')

history = model.fit(
      train_generator,
      steps_per_epoch=55,
      callbacks=[es],
      epochs=100,
      validation_data=validation_generator,
      validation_steps=19)
```

Let's plot our results:

```
# plot the results of the model
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()

plt.figure()

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()
```

The training accuracy and the validation accuracy looks more consistent than the baseline model. Even though the validation accuracy relatively more fluctuates, it keeps increasing along with the training accuracy. We are now able to reach an accuracy of around 83%, a 6% point improvement over the non-regularized model.

```
# evaluate the test images
test_generator = test_datagen.flow_from_directory(
        test_dir,
        target_size=(150, 150),
        batch_size=20,
        class_mode='categorical')

test_loss, test_acc = model.evaluate_generator(test_generator, steps=50)
print('test acc:', test_acc)
```

The test accuracy is nearly 78% which is slightly lower than the validation accuracy but it's still not bad.

We have got the better model compared to the baseline model. However, we can improve the model accuracy more using a pre-trained model to offset the lack of data. Let's build the pre-trained models.

## **Model3: CNN Pretrained Models**

A pre-trained network is simply a saved network previously trained on a large dataset which is general enough so it can be useful for many different computer vision problems. There are two ways to use a pretrained network: feature extraction and fine-tuning. We are going to use both ways with the VGG16 model which is a simple and widely used convnet architecture for Imagenet. 

### Pretrained Model with Feature Extraction

We are going to use the representations learned by a previous network trained on the ImageNet dataset to extract interesting features from our weather images. These features will be used to predict the weather classification.

There are two ways to do feature extraction. Let's build both models. 

#### Feature Extraction without Data Augmentation

We will run the convolutional base over our dataset, record its output to a Numpy array on disk, and then use this data as input to a standalone, densely connected classifier. It only requires running the convolutional base once for every input image but this technique won't allow us to use data augmentation.

Let’s instantiate the VGG16 model.

```
# Instantiating the VGG16 convolutional base
from keras.applications.vgg16 import VGG16

conv_base = VGG16(weights='imagenet', # weight checkpoint to initialize the model from
                  include_top=False, # our dense layer will be trained specific to our small data problem that we are solving
                  input_shape=(150, 150, 3)) # if we don't specify it, the network will be able to process inputs of any size.
```

Here's the detail of the architecture of the VGG16 convolutional base: 

```
conv_base.summary()
```

The final feature map has shape (4, 4, 512) because this is given from the VGG16 architecture. That’s the feature of our weather images on top of which we’ll stick a densely connected classifier.

```
# Extracting features using the pretrained convolutional base
from datetime import datetime
startTime = datetime.now()

# initialize our generator
datagen = ImageDataGenerator(rescale=1./255)
batch_size = 20

def extract_features(directory, sample_count):
    features = np.zeros(shape=(sample_count, 4, 4, 512)) # this is given from the VGG16 architecture!
    labels = np.zeros(shape=(sample_count, 4))
    generator = datagen.flow_from_directory(
        directory,
        target_size=(150, 150),
        batch_size=batch_size,
        class_mode='categorical')
    i = 0
    for inputs_batch, labels_batch in generator:
        features_batch = conv_base.predict(inputs_batch)
        features[i * batch_size : (i + 1) * batch_size] = features_batch
        labels[i * batch_size : (i + 1) * batch_size] = labels_batch
        i += 1
        if i * batch_size >= sample_count:
            # Since generators yield data indefinitely in a loop,
            # we must `break` after every image has been seen once.
            break
    return features, labels

train_features, train_labels = extract_features(train_dir, 1129)
print(datetime.now() - startTime)
validation_features, validation_labels = extract_features(validation_dir, 384)
print(datetime.now() - startTime)
test_features, test_labels = extract_features(test_dir, 378)
print(datetime.now() - startTime)
```

The extracted features currently have (samples, 4, 4, 512) as a shape. We must flatten them to (samples, 8192) to feed them to a densely connected classifier.

```
# These are the X features for modeling
train_features = np.reshape(train_features, (1129, 4 * 4 * 512))
validation_features = np.reshape(validation_features, (384, 4 * 4 * 512))
test_features = np.reshape(test_features, (378, 4 * 4 * 512))
```

Let's define our densely connected classifier with dropout for regularization and train it on the data and labels that we just recorded.

```
# Defining and training the densely connected classifier
model = models.Sequential()
model.add(layers.Dense(256, activation='relu', input_dim=4 * 4 * 512))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(4, activation='softmax'))

model.compile(optimizer=optimizers.RMSprop(lr=2e-5),
              loss='categorical_crossentropy',
              metrics=['acc'])

# early stopping callback
es = EarlyStopping(monitor='val_loss',
                   mode='min',
                   patience=20,
                   restore_best_weights = True)

history = model.fit(train_features, train_labels,
                    callbacks=[es],
                    epochs=100,
                    batch_size=20,
                    validation_data=(validation_features, validation_labels))
```

Let's take a look at the loss and accuracy curves during training:

```
# plot the results of the model
import matplotlib.pyplot as plt

acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()

plt.figure()

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()
```

The validation accuracy of 79% is lower than what we achieve from the CNN with data augmentation model. Also, it seems that the model is overfitting even though we used dropout. This is because the model doesn't leverage data augmentation, which prevents overfitting with our small dataset.

```
# evaluate the test images
test_loss, test_acc = model.evaluate(test_features, test_labels, steps=50)
print('test acc:', test_acc)
```

The test accuracy of 78% is similar to the validation accuracy and the test accuracy of the previous model. 

Now, we are going to run a pretrained model with data augmentation to leverage it during training. 


#### Feature Extraction with Data Augmentation

We will extend the conv_base by adding Dense layers on top, and run the whole thing end to end on our weather images. Every input image will go through the convolutional base every time it's seen by the model. 

Let's add the conv_base to a Sequential model. 

```
# Adding a densely connected classifier on top of the convolutional base
model = models.Sequential()
model.add(conv_base)
model.add(layers.Flatten())
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(4, activation='softmax'))
```

This is what the model looks like now:

```
model.summary()
```

We can see that the convolutional base of VGG16 has 14,714,688 parameters, which is very large and the classifier we added on top has 2 million parameters. This model will take lots of time to run. 

We should freeze the convolutional base to prevent their weights from being updated during training before compiling and training the model. If we don’t do this, then the representations that were previously learned by the convolutional base will be modified during training. 

```
print('This is the number of trainable weights '
      'before freezing the conv base:', len(model.trainable_weights))
```

There are currently 30 trainable weights. Let's freeze the convolutional base.

```
conv_base.trainable = False
```

```
print('This is the number of trainable weights '
      'after freezing the conv base:', len(model.trainable_weights))
```

We can see that there are only 4 trainable weights from the two dense layers that we added. These will be only trained. 

```
# Training the model end to end with a frozen convolutional base
train_datagen = ImageDataGenerator(
      rescale=1./255,
      rotation_range=40,
      width_shift_range=0.2,
      height_shift_range=0.2,
      shear_range=0.2,
      zoom_range=0.2,
      horizontal_flip=True,
      fill_mode='nearest')

# the validation data should not be augmented!
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
        # This is the target directory
        train_dir,
        # All images will be resized to 150x150
        target_size=(150, 150),
        batch_size=20,
        # Since we use categorical_crossentropy loss, we need categorical labels
        class_mode='categorical')

validation_generator = test_datagen.flow_from_directory(
        validation_dir,
        target_size=(150, 150),
        batch_size=20,
        class_mode='categorical')

model.compile(loss='categorical_crossentropy',
              optimizer=optimizers.RMSprop(lr=2e-5),
              metrics=['acc'])

# early stopping callback
from keras.callbacks import EarlyStopping
es = EarlyStopping(monitor='val_loss',
                   mode='min',
                   patience=10,
                   restore_best_weights = True)

history = model.fit(
      train_generator,
      steps_per_epoch=55,
      callbacks=[es],
      epochs=100,
      validation_data=validation_generator,
      validation_steps=19)
```

Let’s plot the results again:

```
# plot the results of the model
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()

plt.figure()

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()
```

We can see that the validation accuracy of 82% is better than previous model and the model seems not overfitting. 

```
test_generator = test_datagen.flow_from_directory(
        test_dir,
        target_size=(150, 150),
        batch_size=20,
        class_mode='categorical')

test_loss, test_acc = model.evaluate_generator(test_generator, steps=50)
print('test acc:', test_acc)
```

The test accuracy of 76% and the validation accuracy of 82% are not better than the simple CNN model with data augmentation. So, let's use fine-tuning which is another widely used technique for model reuse, complementary to feature extraction.

### **Pretrained Model with Fine-tuning**

We are going to unfreeze a few of the top layers of the previous model base used for feature extraction to jointly train these top layers and the fully connected classifier. It will slightly adjust the more abstract representations of the model being reused, in order to make them more relevant for our weather classification at hand. 

As a reminder, this is what our convolutional base looks like:

```
conv_base.summary()
```

We will fine-tune the last three convolutional layers, which means all layers up to block4_pool should be frozen, and the layers block5_conv1, block5_conv2, and block5_conv3 should be trainable. 

The reason that we use only block5 is that layers higher up encode more-specialized features, which are useful to our weather classification. In addition, if we use more layers, it would be risky to attempt to train it on our small dataset, which means the risk of overfitting. 

Let's unfreeze the last three convolutional layers with the below codes.

```
# Freezing all layers up to a specific one
conv_base.trainable = True

set_trainable = False
for layer in conv_base.layers:
    if layer.name == 'block5_conv1':
        set_trainable = True
    if set_trainable:
        layer.trainable = True
    else:
        layer.trainable = False
```

```
# Fine-tuning the model
model.compile(loss='categorical_crossentropy',
              optimizer=optimizers.RMSprop(lr=1e-5),
              metrics=['acc'])

# early stopping callback
es = EarlyStopping(monitor='val_loss',
                   mode='min',
                   patience=10,
                   restore_best_weights = True)

history = model.fit(
      train_generator,
      steps_per_epoch=55,
      callbacks=[es],
      epochs=100,
      validation_data=validation_generator,
      validation_steps=19)
```

Let's plot our results:

```
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()

plt.figure()

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()
```

Even though the model is overfitting, the validation accuracy increased along with the training accuracy. Also, the validation accuracy of 88% is the highest accuracy among the models we've built so far.

```
# evaluate the test images
test_generator = test_datagen.flow_from_directory(
        test_dir,
        target_size=(150, 150),
        batch_size=20,
        class_mode='categorical')

test_loss, test_acc = model.evaluate_generator(test_generator, steps=50)
print('test acc:', test_acc)
```

The test accuracy of 82% is also the highest among the models we've built so far.
